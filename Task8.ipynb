```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Espri metinlerini vektörleştirmek için TfidfVectorizer kullanma
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(dfjoke['JokeText'])

# Kosinüs benzerliğini hesapla
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
titles = dfjoke['JokeText']
indices = pd.Series(dfjoke.index, index=dfjoke['JokeText'])

def get_recommendations(joke_text):
    idx = indices[joke_text]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:31]
    joke_indices = [i[0] for i in sim_scores]
    return titles.iloc[joke_indices]

example_joke = dfjoke['JokeText'].iloc[0]
recommended_jokes = get_recommendations(example_joke).head(10)

recommended_jokes
```

```python
!pip3 install scikit-surprise #or !conda install -c conda-forge scikit-surprise
from surprise import Dataset
from surprise import Reader
from surprise.model_selection import train_test_split
```

```python
df_long = dfrtng.melt(id_vars=['JokeId'], var_name='User', value_name='Rating')

df_long = df_long.dropna(subset=['Rating'])

df_long['JokeId'] = df_long['JokeId'].astype(str)
df_long['User'] = df_long['User'].astype(str)

reader = Reader(rating_scale=(0, 5))
X = Dataset.load_from_df(df_long[['User', 'JokeId', 'Rating']], reader)

X_train, X_test = train_test_split(X, test_size=.25)

print(df_long)
```

```python
from surprise import SVD

mdlSvdrtng = SVD()
```

```python
mdlSvdrtng.fit(X_train)
test_pred = mdlSvdrtng.test(X_test)
```

```python
from surprise import accuracy

accuracy.rmse(test_pred)
```

```python
from surprise.model_selection import GridSearchCV

param_grid = {'n_epochs': [5, 10, 15], 'lr_all': [0.002, 0.005],
              'reg_all': [0.4, 0.6]}
gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=3)

gs.fit(X)

print(gs.best_score['rmse'])

print(gs.best_params['rmse'])
```

```python
from surprise.model_selection import cross_validate

cross_validate(mdlSvdrtng, X, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```
