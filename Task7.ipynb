```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
%%time
# Speed up by using pattern fragment growth method for mining frequent itemsets (uses density to find good itemsets)

from mlxtend.frequent_patterns import fpgrowth

frequent_itemsets = fpgrowth(dftrb, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
te = TransactionEncoder()
te_array = te.fit(dftrList).transform(dftrList)
dftrb = pd.DataFrame(te_array, columns=te.columns_)

# Check the transformed DataFrame
print("Sample of transformed DataFrame:", dftrb.head())

# Apply the apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(dftrb, min_support=0.01, use_colnames=True)

# Check the frequent itemsets
print("Frequent itemsets:", frequent_itemsets)

# If frequent_itemsets is empty, lower the min_support
if frequent_itemsets.empty:
    print("No frequent itemsets found. Trying a lower min_support...")
    frequent_itemsets = apriori(dftrb, min_support=0.001, use_colnames=True)
    print("Frequent itemsets with lower min_support:", frequent_itemsets)

# Generate the association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Check the generated rules
print("Generated rules:", rules)

# Filter the rules with at least 70% confidence
high_confidence_rules = rules[rules['confidence'] >= 0.7]

# Display the high confidence rules
print("High confidence rules:", high_confidence_rules)

# Ensure there are rules to plot
if not high_confidence_rules.empty:
    # Optionally, visualize the high confidence rules
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='support', y='confidence', size='lift', hue='lift', data=high_confidence_rules, palette='viridis', legend=False, sizes=(20, 200))
    plt.title('Association Rules with at least 70% Confidence')
    plt.xlabel('Support')
    plt.ylabel('Confidence')
    plt.show()
else:
    print("No association rules with at least 70% confidence were found.")
```
