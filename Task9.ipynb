```python
# Use pipelines to access pre-trained models
from transformers import pipeline
```

## Sentiment Analysis

```python
# Model
mdlSnt = pipeline('sentiment-analysis')

# Parameters
#prmStatement = 'We are happy to go on vacation this spring break.'
prmStatement = ['It is nice to have you here in the US.  I heard the weather is terrible in the UK.',
        'I am struggling with stress']
# Predict
mdlSnt(prmStatement)
```

Can perform sentiment analysis in other languages as well using `model = 'nlptown/bert-base-multilingual-uncased-sentiment'`

```python
# Model
mdlSnt = pipeline('sentiment-analysis', model = 'nlptown/bert-base-multilingual-uncased-sentiment')

# Parameters
prmStatement = 'hace mucho calor' #spanish for: the weather is very hot

# Predict
mdlSnt(prmStatement)
```

The result is in stars as opposed to score... this is as per the model definition.  See model details in HF for more details.

## Question Answering

### Example 1

```python
from google.colab import drive
drive.mount('/content/drive')

mdlQa = pipeline("question-answering")

# Parameters
f = open("/content/drive/MyDrive/Colab Notebooks/SpaceX.txt", "r")
context = f.read()
new_text_content = """
SpaceX, officially known as Space Exploration Technologies Corp., is an American aerospace manufacturer and space transportation company founded by Elon Musk in 2002.
The company is known for developing the Falcon 1, Falcon 9, and Falcon Heavy launch vehicles, along with the Dragon spacecraft.
SpaceX was the first privately-funded company to send a spacecraft to the International Space Station in 2012.
Their Starship project aims to develop a fully reusable spacecraft for missions to Mars and beyond.
SpaceX has achieved numerous milestones, including the first reuse of an orbital rocket and the first private crewed mission to orbit in 2020.
In May 2020, SpaceX became the first private company to send astronauts to the International Space Station with NASA's Crew Dragon Demo-2 mission.
The company has revolutionized the space industry by significantly reducing the cost of access to space, making it possible to achieve ambitious goals such as the colonization of Mars.
SpaceX's long-term vision includes enabling human life on other planets and becoming a multi-planetary species.
"""

# Create a new .txt file with the content
with open("spacex.txt", "w") as file:
    file.write(new_text_content)

# Read the new .txt file
with open("spacex.txt", "r") as file:
    context = file.read()

# Use the context with the question-answering model
from transformers import pipeline

# Model
mdlQa = pipeline("question-answering")

# Parameters
question = "Who founded SpaceX?"

# Predict
answer = mdlQa(question=question, context=context)

# Print the result
print("Answer:", answer["answer"])
```
